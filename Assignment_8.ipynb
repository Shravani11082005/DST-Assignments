```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs

# Generate synthetic unlabeled dataset
X, _ = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=0)

# Elbow Method to find optimal k
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot Elbow curve
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid()
plt.show()

# Fit KMeans with chosen k (e.g., 5 from elbow)
kmeans = KMeans(n_clusters=5, random_state=42)
labels = kmeans.fit_predict(X)

# Visualize clusters using PCA (2D projection)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
plt.title('KMeans Clustering (PCA Projected)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid()
plt.show()
```

```python
from sklearn.datasets import load_digits

# Load dataset
digits = load_digits()
X_digits = digits.data
y_digits = digits.target

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_digits)

# Visualize PCA result
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_digits, cmap='tab10', alpha=0.7)
plt.legend(*scatter.legend_elements(), title="Digits", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('PCA on Digits Dataset')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid()
plt.tight_layout()
plt.show()
```
